```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This is just for demonstration purposes to show you what different degree curves look like
#Pay special attention to the values on the Y-axis
x = np.linspace(1,6,1000)
for i in range(1,6):
    plt.plot(x,x**i,label = i)
    plt.legend(loc = "best")
    plt.show()

plt.plot(x,x**8,label = 8)
plt.legend(loc = "best")
plt.show()
```

```python
#uncomment each matrix and comment all others and run subsequent line of code to see the output change
#All the below matrices contain only one features, ie X1
#The difference is that first one contains only X1
#second one contains X1 and X1**2
#Third contains X1,X1**2 and X1**3 and so on...
y = np.matrix([[5],[7],[10],[12],[20]])
#ORIGINAL DATASET - [[1],[2],[3],[4],[5]]
#x = np.matrix([[1,1],[1,2],[1,3],[1,4],[1,5]])
#x = np.matrix([[1,1,1],[1,2,4],[1,3,9],[1,4,16],[1,5,25]])
#x = np.matrix([[1,1,1,1],[1,2,4,8],[1,3,9,27],[1,4,16,64],[1,5,25,125]])
x = np.matrix([[1,1,1,1,1],[1,2,4,8,16],[1,3,9,27,81],[1,4,16,64,256],[1,5,25,125,625]])
#x = np.matrix([[1,1,1,1,1,1],[1,2,4,8,16,32],[1,3,9,27,81,243],[1,4,16,64,256,1024],[1,5,25,125,625,3125]])
```

```python
#This whole process remains the same as Multilinear Regression, Refer that notebook for explanation
xt = x.T
xtx = np.dot(xt,x)
inv_xtx = xtx.I
theta = np.dot(inv_xtx,np.dot(xt,y))
```

```python
print(x)
theta
```

```python
pred_y = np.dot(x,theta)
np.array(pred_y)[:,0]
```

```python
#As you can notice as we increase degree of X1 in our above toy dataset, the degree of curve changes accordingly
#This is the main difference in multilinear regression (MLR) and polynomial regression (PR)
#in MLR, A plane would form on taking 2 features as input but in PR, you can increase degree of curve to find a better fit
#You can combine MLR and PR to get shapes such as paraboloid (Take 2 features and take both of them to the second degree)
#The only thing that will change is the yhat equation
pred_y = np.dot(x,theta)
plt.scatter(np.array(x[:,1]),np.array(y[:,0]),
            color = 'blue')
plt.plot(x[:,1],pred_y[:,0],color = 'red')
plt.scatter([l[0] for l in x[:,1].tolist()],
            [r[0] for r in pred_y[:,0].tolist()],
            color = 'red')
#plt.scatter([3.2],[12],color='k')
plt.show()
```
